```python
from nltk.corpus import brown, gutenberg
import pprint
from gensim.test.utils import datapath, get_tmpfile
from gensim.models import KeyedVectors
from gensim.scripts.glove2word2vec import glove2word2vec
import pickle
```

```python
glove_path = '/Users/yilingc/Courses/CS398/project/data/glove.840B.300d.txt'
```

```python
glove_file = datapath(glove_path)
word2vec_glove_file = get_tmpfile("/Users/yilingc/Courses/CS398/project/data/glove.840B.300d.word2vec.txt")
glove2word2vec(glove_file, word2vec_glove_file)
model = KeyedVectors.load_word2vec_format(word2vec_glove_file)
```

```python
glove_words = set()
with open(glove_path) as f:
    for line in f:
        glove_words.add(line.rsplit()[0])
```

```python
pickle.dump(glove_words, open('data/glove_words.pickle', 'wb'))
```
